```python
# 3. another for loop to get channel info and links for each channel from the excel file `Link` column
# Function to extract channel information and links from yt.lemnoslife.com
def get_channel_info(channel_id):
    url = f"https://yt.lemnoslife.com/channels?part=about&id={channel_id}"
    try:
        response = requests.get(url)
        data = response.json()
        if 'items' in data:
            channel_info = data['items'][0]
            about_links = {}
            if 'about' in channel_info and 'links' in channel_info['about']:
                for link in channel_info['about']['links']:
                    about_links[link['title']] = link['url']
            return channel_info, about_links
        else:
            return None, None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None, None

# Read Excel file containing channel information
excel_file = "BGM_list_update.xlsx"  # Replace with your file path
df = pd.read_excel(excel_file)

# Iterate through each row in the DataFrame
for index, row in tqdm(df.iterrows()):
    channel_url = row['Link']
    
    # Check if channel_url is a string
    if isinstance(channel_url, str):
        # Extract channel ID from channel URL
        channelID = channel_url.split('/')[-1]

        # Retrieve channel information and links
        channel_info, about_links = get_channel_info(channelID)

        # Update the DataFrame with channel information and links
        if channel_info:
            df.at[index, ['subscriberCount', 'videoCount', 'viewCount', 'location']] = [channel_info.get('follower'), channel_info.get('video'), channel_info.get('view'), channel_info.get('country')]
            if about_links:
                df.at[index, ['Email', 'Website', 'Facebook', 'Instagram', 'TikTok', 'Twitter']] = [about_links.get('Email'), about_links.get('Website'), about_links.get('FB'), about_links.get('IG'), about_links.get('Tiktok'), about_links.get('Twitter')]
            else:
                df.at[index, ['Email', 'Website', 'Facebook', 'Instagram', 'TikTok', 'Twitter']] = None
        else:
            df.iloc[index, :] = None  # Set all columns in the row to None
    else:
        # Handle the case where channel_url is not a string (e.g., NaN)
        print(f"Invalid channel URL at index {index}: {channel_url}")
```
